```python

import plotly
from plotly.graph_objs import graph_objs as go
from plotly import tools
#from ipywidgets import interactive
import ipywidgets as widgets
import numpy as np
from IPython.display import display, Math, Latex, clear_output
plotly.offline.init_notebook_mode()
from plotly.offline import iplot
import time
print("SDF")
```

```python
# USE THIS LINK https://plot.ly/python/subplots/ AND OVERLAP SUBPLOTS THEN JUST UPDATE THE PLOT WITH 
# THE FUNCTION ON IT


class functions:
    
    def __init__(self, functions):
        self.functions = functions
   
    def parabola(self, y, c, b, a):
        return a*(y + b)**2  +c  
    
    def log_10(self, y, c, b, a):
        return a*np.log10(y+b) + c

    def cubic(self, y, c, b,a):
        return a*(y + b)**3 + c
    
    def sine(self, y, c, b, a):
        return a*np.sin(y+b) + c
    
    def tan(self, y, c, b, a):
        return a*np.tan(y+b) + c
    
    def run_func(self, func_key, *args, **kwargs):
        func_name = self.functions.get(func_key)
        if func_name and hasattr(self, func_name):
            return getattr(self, func_name)(*args, **kwargs)
        else:
            return None


Dictionary = {
         'parabola':'parabola',
         'log_10':'log_10',
         'cubic':'cubic',
         'sine':'sine',
         'tan':'tan'
         }

titles = {
        "parabola" : '$a(x^2 + b) + c$',
        "log_10": "$a\log_{10}(x + b) + c$",
        "cubic" : "$a(x+b)^3 + c$",
        "sine" : "$a \sin(x + b) + c$",
        "tan": "$a \\tan(x+b) + c$",
        }

CallFunctions = functions(Dictionary)
layout = go.Layout(
    xaxis1=dict(
        autorange=False,
        showgrid=True,
        zeroline=True,
        showline=True,
        autotick=True,
        ticks='',
        showticklabels=True,
        range=[-10, 10],
        title = "x",
        domain = [0,1]
    ),
    yaxis1=dict(
        autorange=False,
        showgrid=True,
        zeroline=True,
        showline=True,
        autotick=True,
        ticks='',
        showticklabels=True,
        range=[-10, 10],
        title = "y",
        domain = [0,1]
    )
 
    )

# fig = go.Figure(data=[{'x':None, 'y':None, 'yaxis' :'yaxis', 'xaxis':'xaxis'}], layout=layout)
fig = tools.make_subplots(rows=1, cols=2, shared_yaxes=True, shared_xaxes = True)
fig['layout'].update(layout)
#fig['layout']['updatemenus'] = [{'redraw':False}]
fig.append_trace(go.Scatter(x = [], y = []),1,1)
fig.append_trace(go.Scatter(x = [], y = []),1,1)



def View(a, b, c, function):
    
    x1 = np.linspace(-10,10,500)
    y = CallFunctions.run_func(function, x1, c, b, a)
   
    
    trace = go.Scatter(x = list(x1), 
                     y = y,
                     name = function,
                     )      
    
    fig['data'][1]= trace

    plotly.offline.iplot(fig)
                     
import warnings
import sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
    
interactive_plot = widgets.interactive(View, 
                               c=(-8,8,.5), 
                               b = (-8,8,.1), 
                               a = (0.005, 10,.1),
                               function=Dictionary, 
                               continuous_update=True,
                               wait=True)

output = interactive_plot.children[-1]
output.layout.height = '600px'
output.clear_output(wait=True) 
interactive_plot 



    
```
