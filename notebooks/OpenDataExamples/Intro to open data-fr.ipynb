# Introduction to Open Data




## What is "Open Data"?
Open data is data that is freely available to use, manipulate, change, redistribute and analyze without restriction. Open data can be easily accessed by anyone, anywhere, anytime provided they possess the means to access that data via the Internet.  There is mounting pressure around the world for both governments and private entities to make more data available for analysis (and therefore scrutiny) by the public. The idea behind these initiatives is that to will allow for greater transparency from governments and businesses with the general public. As more data is becoming available, learning how to find and work with that data is becoming ever more important, and in our "Open Data 101" tutorial series, we'll introduce you to some of the basic tools and techniques you can use in a Jupyter notebook to load, analyze and interpret the results of your analysis of open data. 


## Are Some Datasets Better Than Others? 

Objectively speaking, the simple answer to this is 'yes'. Regardless of the question you're trying to answer or the analysis you're trying to do - just because data is open, doesn't necessarily mean that it is going to be easy to work with! Here's a few general rules to look for to make sure the data set you've chosen will be easy to manipulate and work with. 

1. If the data is in multiple files, is it consistently labeled/easy to relate quantities between files?  

1. Is the data in a format that's easy to load on your computer such as `csv`, `txt`, `json`, or Excel?   
 1. A kindly formatted website HTML table is also an excellent format!  

1. If you're working with spacial data, are GPS coordinates already included (where relevant)?   
1. Is the dataset complete for your problem? If values are missing, is there a good reason for them to be missing and are they very clearly labelled as missing?


For certain projects the maintenance of a data set (is it updated as new data become available?) is important. Typically there will be some mention of if the data set is maintained on the download page, or at the very least a time stamp of the last time the dataset was updated. 


### Reliable Sources
Another problem with open data is some data sets may be considered to be dubious, or constructed in a way that is intentionally misleading or simply incorrect. 
Much like any other source of information, a data set may also be affected by any biases the curator of the data may have (as well as your own!). Because of this, it is important to try and find an unbiased source of data, or at the very least multiple sources of data which may have alternate biases. For example, if you're looking to find data in order to study the heath affects of tobacco, a dataset provided by a study which was funded by a tobacco company may have entirely different results than one which was done independently. As this is the case for all data sets, it is important to be conscious of your data's origins before you draw any major conclusions from your analysis. 

# Introduction à l'open data




## Qu'est-ce que "Open Data"?
Les données ouvertes sont des données librement disponibles pour être utilisées, manipulées, modifiées, redistribuées et analysées sans restriction. Les données ouvertes sont facilement accessibles par tous, n’importe où et à tout moment, à condition qu’ils possèdent les moyens d’accéder à ces données via Internet. Les gouvernements et les entités privées exercent de plus en plus de pression dans le monde entier pour que davantage de données soient disponibles pour l'analyse (et donc le contrôle) par le public. L’idée qui sous-tend ces initiatives est de permettre une plus grande transparence des gouvernements et des entreprises vis-à-vis du grand public. À mesure que de plus en plus de données sont disponibles, il est de plus en plus important d’apprendre à les trouver et à les utiliser. Dans notre série de didacticiels «Open Data 101», nous vous présenterons quelques-uns des outils et techniques de base que vous pouvez utiliser. un bloc-notes Jupyter pour charger, analyser et interpréter les résultats de votre analyse de données ouvertes.


## Certains ensembles de données sont-ils meilleurs que d’autres?

Objectivement, la réponse simple à cette question est «oui». Quelle que soit la question à laquelle vous essayez de répondre ou l'analyse que vous tentez de faire, le simple fait que les données soient ouvertes ne signifie pas nécessairement qu'il sera facile de travailler avec! Voici quelques règles générales à suivre pour vous assurer que le jeu de données que vous avez choisi sera facile à manipuler et à utiliser.

1. Si les données sont dans plusieurs fichiers, est-il systématiquement étiqueté / facile de relier les quantités entre les fichiers?

1. Les données sont-elles dans un format facile à charger sur votre ordinateur, tel que `csv`,` txt`, `json` ou Excel?
 1. Un tableau HTML de site Web bien formaté est également un excellent format!

1. Si vous travaillez avec des données spatiales, les coordonnées GPS sont-elles déjà incluses (le cas échéant)?
1. Le jeu de données est-il complet pour votre problème? Si des valeurs manquent, existe-t-il une bonne raison de les manquer et sont-elles très clairement étiquetées comme manquantes?


Pour certains projets, la maintenance d'un ensemble de données (est-il mis à jour lorsque de nouvelles données deviennent disponibles?) Est importante. Généralement, il sera mentionné si le jeu de données est maintenu sur la page de téléchargement, ou au moins un horodatage de la dernière mise à jour du jeu de données.


### Sources fiables
Un autre problème avec les données ouvertes est que certains ensembles de données peuvent être considérés comme douteux ou construits de manière intentionnellement trompeuse ou tout simplement incorrecte.
Comme toute autre source d’information, un ensemble de données peut également être affecté par un biais éventuel du conservateur des données (ainsi que par vous-même!). Pour cette raison, il est important d'essayer de trouver une source de données non biaisée, ou au moins de multiples sources de données pouvant avoir des biais différents. Par exemple, si vous recherchez des données pour étudier les effets néfastes du tabac sur la santé, un ensemble de données fourni par une étude financée par une société de tabac peut donner des résultats totalement différents de ceux obtenus indépendamment. Comme c'est le cas pour tous les ensembles de données, il est important de connaître les origines de vos données avant de tirer des conclusions majeures de votre analyse.
